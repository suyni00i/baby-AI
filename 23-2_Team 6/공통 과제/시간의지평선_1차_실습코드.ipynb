# **1.데이터시각화**

### **1.데이터시각화 (1.Matplotlib)**
##### 대상기업 : 맥도날드(code: MCD) 웹사이트 : 야후 파이낸스

```python
# 1.웹 스크래핑 - 야후파이낸스, 맥도날드(코드: MCD)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# 2.데이터 수집 - 해당 주식의 1년 이상 일일 종가(Close) 데이터 수집
start = '2022-01-01'
end = '2023-08-01'
symbol = 'MCD'

df = pd.DataFrame(yf.download(symbol, start, end))
df
```

```python
# 3. Matplotlib 으로 주식의 가격 변동 시각화하기

plt.plot(df['Close'], label='Close Price')
plt.title('McDonalds(2022-01-01 to 2023-07-31)')
plt.legend(loc='upper left')

plt.xlabel('Date')
plt.xticks(rotation=45)        # X축에 들어갈 내용의 기울기 조정
plt.ylabel('Price(USD)')

plt.show();
```

```python
# 4.30일/90일 이동평균 구하기
df['MA30'] = df['Close'].rolling(30).mean()
df['MA90'] = df['Close'].rolling(90).mean()
df
```

```python
# 5.그래프 확장

plt.title('McDonalds(2022-01-01 to 2023-07-31)')
plt.xlabel('Date')
plt.xticks(rotation=45)        # X축에 들어갈 내용의 기울기 조정
plt.ylabel('Price(USD)')

plt.plot(df['Close'], label='Close Price', color='black', linewidth='1')
plt.plot(df['MA30'], label='M.A 30Days', color='blue', linewidth='2')
plt.plot(df['MA90'], label='M.A 90Days', color='red', linewidth='2')

plt.legend(loc='upper left', fontsize=14, frameon=True, shadow=True)
plt.show();
```

# **2.사칙연산**

문제
각 사칙연산에 해당하는 총 4개의 함수를 작성하고, 정수의 연산식이 주어지면 적절한 함수를 호출하여 4칙연산의 연산 결과를 출력하는 프로그램을 작성해보세요.
 단, '/'의 결과는 정수 부분만 출력하고, 사칙연산자 이외의 문자가 주어지는 경우에는 False 을 출력합니다.

```python
def x(a, b, o):
  if o == '+':
    print(f'{a} + {b} = {a + b}')
  elif o == '-':
    print(f'{a} - {b} = {a - b}')
  elif o == '*':
    print(f'{a} * {b} = {a * b}')
  elif o == '/':
    print(f'{a} / {b} = {a // b}')
  else:
    print("False")

a, o, b = input().split()
a = int(a)
b = int(b)

x(a, b, o)
```
