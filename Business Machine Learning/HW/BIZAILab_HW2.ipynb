```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('bikeshare_pre.csv')
```

## Polynomial Regression

```python
A = data.drop('count', axis=1, inplace=False)
b = data['count']
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=True)
```

```python
A_poly = poly.fit_transform(A)
Ap_train, Ap_test, bp_train, bp_test = train_test_split(A_poly, b, test_size=0.3, random_state=123)
```

## Lasso Regularization

```python
# alpha_values = [0.0001, 0.0002, 0.0003, 0.0004, 0.0005]
alpha_values = [1e-9, 1e-07, 1e-05, 1e-03, 1e-01]
rmse_values = []
```

```python
for alpha in alpha_values:
    lassopr = Lasso(alpha=alpha)
    lassopr.fit(Ap_train, bp_train)

    bptrain_result = lassopr.predict(Ap_train)
    bptest_result = lassopr.predict(Ap_test)

    rmse_train = np.sqrt(mean_squared_error(bp_train, bptrain_result))
    rmse_test = np.sqrt(mean_squared_error(bp_test, bptest_result))

    rmse_values.append((alpha, rmse_train, rmse_test))
```

```python
for alpha, rmse_train, rmse_test in rmse_values:
    print(f'Alpha = {alpha}:')
    print('Train 데이터의 RMSE 값 {}'.format(rmse_train))
    print('Test  데이터의 RMSE 값 {}\n'.format(rmse_test))
```

### 최적의 alpha

```python
lassopr = Lasso(alpha=1e-5)

lassopr.fit(Ap_train, bp_train)

bptrain_result = lassopr.predict(Ap_train)
bptest_result  = lassopr.predict(Ap_test)

rmse_train = np.sqrt(mean_squared_error(bp_train, bptrain_result))
rmse_test = np.sqrt(mean_squared_error(bp_test, bptest_result))

print('Train 데이터의 RMSE 값 {}'.format(rmse_train))
print('Test  데이터의 RMSE 값 {}'.format(rmse_test))
```
