# 1. LinearRegression 모듈 이용 계산

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
```

```python
HW = pd.DataFrame({'SNS 시간': [10,80, 50, 40], '직전시험점수':[95,20,60,50], '시험점수':[90,10,85,50]})
HW
```

```python
X = HW.drop("시험점수", axis = 1, inplace = False)
y = HW["시험점수"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
coef = pd.Series(data = model.coef_, index = X.columns)
print(coef)
```

# 2. 직접 계산

```python
import pandas as pd
import numpy as np
from numpy.linalg import inv
from numpy.linalg import pinv
```

```python
HW = pd.DataFrame({'SNS 시간': [10,80, 50, 40], '직전시험점수':[95,20,60,50], '시험점수':[90,10,85,50]})
HW
```

```python
X = HW.drop("시험점수", axis = 1, inplace = False)
y = HW["시험점수"]
```

```python
# Dataframe to numpy
X = pd.DataFrame(X).to_numpy()
y = pd.DataFrame(y).to_numpy()
```

```python
# inv
Xpinv = inv(X.T @ X) @ X.T
print(Xpinv)

# pseufo inv
Xpinv2 = pinv(X)
print(Xpinv2)
```

```python
x = Xpinv @ y

np.set_printoptions(precision=8, suppress=True)
print(x)
```
